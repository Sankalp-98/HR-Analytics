```python
df_train_dummy=pd.get_dummies(df_train,drop_first=True)
df_train_dummy
```

```python
df_test_dummy=pd.get_dummies(df_test,drop_first=True)
df_test_dummy
```

```python
X=df_train_dummy.drop(columns=['employee_id','is_promoted'])
y=df_train_dummy['is_promoted']
```

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
X_new=pd.DataFrame(sc.fit_transform(X))
X_new.columns=X.columns
X_new.index=X.index

X_new
```

```python
X_test=df_test_dummy.drop(columns=['employee_id'])
```

```python
sc=MinMaxScaler()
X_test_new=pd.DataFrame(sc.fit_transform(X_test))
X_test_new.columns=X_test.columns
X_test_new.index=X_test.index

X_test_new
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X_new,y,random_state=21,test_size=0.2,stratify=y)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model=LogisticRegression()
```

```python
lg_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import f1_score
pred=lg_model.predict(X_valid)
f1_score(pred,y_valid)

```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=60,n_jobs=-1,oob_score=True,min_samples_split=10)
```

```python
rf_model.fit(X_train,y_train)
```

```python
pred=rf_model.predict(X_valid)
f1_score(pred,y_valid)
```

# XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier(learning_rate=0.1, n_estimators=200, max_depth=4, min_child_weight=7, 
                      gamma=0.4,nthread=4, subsample=0.8, colsample_bytree=0.8, objective= 'binary:logistic',scale_pos_weight=3,seed=29)
```
